## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

**Minh họa BFS**:
1. Khởi tạo: Hàng đợi = [(S, [S])], Đã thăm = {S}
2. Lấy S, thêm A, B, C: Hàng đợi = [(A, [S, A]), (B, [S, B]), (C, [S, C])], Đã thăm = {S, A, B, C}
3. Lấy A, thêm D: Hàng đợi = [(B, [S, B]), (C, [S, C]), (D, [S, A, D])], Đã thăm = {S, A, B, C, D}

1. Khởi tạo: Ngăn xếp = [(S, [S])], Đã thăm = {S}

3. Lấy A, thêm D: Ngăn xếp = [(D, [S, A, D])], Đã thăm = {S, A, D}

5. Lấy F, không có kề mới: Ngăn xếp rỗng, quay lại S.
6. Từ S, thêm B: Ngăn xếp = [(B, [S, B])], Đã thăm = {S, A, D, F, B}
7. Lấy B, thêm E: Ngăn xếp = [(E, [S, B, E])], Đã thăm = {S, A, D, F, B, E}
8. Lấy E, thêm G: G là đích, trả về [S, B, E, G]

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

BFS trên đồ thị không trọng số:

Mỗi cạnh được xem như bằng nhau (trọng số = 1).
→ BFS đi theo từng “lớp” (level), nên luôn tìm ra đường đi ít cạnh nhất từ đỉnh bắt đầu đến đỉnh đích.
→ Đây chính là đường đi ngắn nhất.

BFS trên đồ thị có trọng số:

BFS không quan tâm trọng số các cạnh → chỉ xét đường đi nào đến trước.

→ Vì vậy có thể tìm ra đường đi sớm nhất, nhưng chưa chắc đường có tổng trọng số nhỏ nhất.

Ví dụ minh họa từ Đồ thị mẫu 5:
Có 3 đường từ S → G:
S → A → G: trọng số = 1 + 3 = 4 (BFS chọn đường này vì đến trước)

S → B → G: trọng số = 2 + 5 = 7

S → C → D → G: trọng số = 3 + 6 = 9

→ BFS chọn S → A → G vì đi qua ít đỉnh nhất, nhưng không đảm bảo là tổng trọng số nhỏ nhất.

Kết luận:

BFS chỉ đảm bảo tìm đường ngắn nhất theo số bước nếu đồ thị không trọng số.

Nếu đồ thị có trọng số, thì cần dùng thuật toán Dijkstra để tìm đường đi có tổng trọng số nhỏ nhất (tối ưu nhất).

2.

Ví dụ: A → B → A → B → ...

Cách tránh:

→ Dùng một tập đã thăm (visited) để đánh dấu những đỉnh đã đi qua.

Với BFS: khi đưa đỉnh vào hàng đợi thì thêm vào visited.

Với DFS: khi bắt đầu đệ quy vào đỉnh đó, đánh dấu ngay.

Áp dụng vào Đồ thị mẫu 2:

Nếu không đánh dấu B đã thăm thì thuật toán có thể lặp vòng giữa các nhánh.

→ Kết luận: Phải dùng visited để tránh đi lại các đỉnh đã thăm → tránh lặp vô hạn.

````python
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque

def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set([start])
    
    while queue:
        (node, path) = queue.popleft()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None

# Đồ thị mẫu
graph2 = {
    'S': ['A', 'B'],
    'A': ['S', 'C', 'B'],
    'B': ['S', 'A', 'D'],
    'C': ['A', 'D'],
    'D': ['C', 'B', 'G'],
    'G': ['D']
}

# Thực hiện BFS
path = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)

# Vẽ đồ thị bằng networkx
G = nx.Graph()

# Thêm các cạnh vào đồ thị
for node in graph2:
    for neighbor in graph2[node]:
        G.add_edge(node, neighbor)

# Tạo layout
pos = nx.spring_layout(G)

# Vẽ các node và cạnh
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=14)

# Nếu tìm được đường đi thì highlight nó
if path:
    # Tạo danh sách các cạnh trong đường đi
    path_edges = list(zip(path, path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, width=4, edge_color='red')
    nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='orange', node_size=1200)

plt.title("Đường đi BFS từ {} đến {}".format(path[0], path[-1]))
plt.show()
````

### Bài Tập Nâng Cao

2. Thiết kế một đồ thị có trọng số với ít nhất 10 nút và 15 cạnh. Chạy BFS và DFS, phân tích đường đi và tổng trọng số. Đảm bảo mã có chú thích chi tiết.

````python
# Đồ thị mẫu 6
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
import time

# Khởi tạo danh sách cạnh có trọng số
edges = [
    ('S', 'A', 2), ('S', 'C', 5), ('A', 'B', 3),
    ('A', 'D', 4), ('B', 'E', 6), ('C', 'D', 7),
    ('C', 'F', 9), ('D', 'E', 8), ('E', 'H', 10),
    ('F', 'G', 12), ('G', 'H', 15)
]

# Tạo graph không trọng số để BFS / DFS
graph = {}
for u, v, _ in edges:
    graph.setdefault(u, []).append(v)
    graph.setdefault(v, []).append(u)

# Hàm BFS và DFS
def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set([start])
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None

def dfs(graph, start, goal):
    stack = [(start, [start])]
    visited = set([start])
    while stack:
        node, path = stack.pop()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append((neighbor, path + [neighbor]))
    return None

# Tính tổng trọng số đường đi
def calculate_weight(path, edge_list):
    edge_weight = {(u, v): w for u, v, w in edge_list}
    edge_weight.update({(v, u): w for u, v, w in edge_list})  # vì undirected
    total = 0
    for i in range(len(path) - 1):
        total += edge_weight[(path[i], path[i+1])]
    return total

# Chạy BFS và DFS từ S đến H
start, end = 'S', 'H'

t1 = time.time()
bfs_path = bfs(graph, start, end)
t2 = time.time()

t3 = time.time()
dfs_path = dfs(graph, start, end)
t4 = time.time()

# In kết quả
print("BFS path:", bfs_path)
print("BFS time:", round(t2 - t1, 6), "s")
if bfs_path:
    print("BFS total weight:", calculate_weight(bfs_path, edges))

print("\nDFS path:", dfs_path)
print("DFS time:", round(t4 - t3, 6), "s")
if dfs_path:
    print("DFS total weight:", calculate_weight(dfs_path, edges))

# Vẽ đồ thị với trọng số và đường đi
G = nx.Graph()
for u, v, w in edges:
    G.add_edge(u, v, weight=w)

pos = nx.spring_layout(G, seed=42)
edge_labels = nx.get_edge_attributes(G, 'weight')

plt.figure(figsize=(12, 6))

# Vẽ BFS
plt.subplot(1, 2, 1)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
if bfs_path:
    bfs_edges = list(zip(bfs_path, bfs_path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=bfs_edges, edge_color='red', width=3)
    nx.draw_networkx_nodes(G, pos, nodelist=bfs_path, node_color='orange', node_size=1200)
plt.title("Đường đi BFS từ S đến H")

# Vẽ DFS
plt.subplot(1, 2, 2)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
if dfs_path:
    dfs_edges = list(zip(dfs_path, dfs_path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=dfs_edges, edge_color='purple', width=3)
    nx.draw_networkx_nodes(G, pos, nodelist=dfs_path, node_color='orange', node_size=1200)
plt.title("Đường đi DFS từ S đến H")

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
import time

# Tạo danh sách cạnh có trọng số
edges = [
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'D', 5),
    ('C', 'D', 1), ('C', 'E', 8), ('D', 'F', 3),
    ('E', 'F', 2), ('E', 'G', 4), ('F', 'H', 6),
    ('G', 'H', 1), ('G', 'I', 7), ('H', 'J', 5),
    ('I', 'J', 3), ('B', 'G', 9), ('D', 'I', 6)
]

# Tạo graph không trọng số cho BFS & DFS
graph = {}
weights = {}

for u, v, w in edges:
    graph.setdefault(u, []).append(v)
    graph.setdefault(v, []).append(u)
    weights[(u, v)] = w
    weights[(v, u)] = w  # đồ thị vô hướng

# Hàm BFS trả về đường đi
def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set([start])

    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None

# Hàm DFS trả về đường đi
def dfs(graph, start, goal):
    stack = [(start, [start])]
    visited = set([start])

    while stack:
        node, path = stack.pop()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append((neighbor, path + [neighbor]))
    return None

# Tính tổng trọng số của đường đi
def total_weight(path, weights):
    return sum(weights[(u, v)] for u, v in zip(path, path[1:]))

# Chạy BFS và DFS và đo thời gian 
start_node, end_node = 'A', 'J'

t1 = time.time()
bfs_path = bfs(graph, start_node, end_node)
t2 = time.time()

t3 = time.time()
dfs_path = dfs(graph, start_node, end_node)
t4 = time.time()

# In kết quả
print("=== BFS ===")
print("Đường đi:", bfs_path)
print("Tổng trọng số:", total_weight(bfs_path, weights))
print("Thời gian:", round(t2 - t1, 6), "giây")

print("\n=== DFS ===")
print("Đường đi:", dfs_path)
print("Tổng trọng số:", total_weight(dfs_path, weights))
print("Thời gian:", round(t4 - t3, 6), "giây")

# Vẽ đồ thị và đánh dấu đường đi
G = nx.Graph()
for u, v, w in edges:
    G.add_edge(u, v, weight=w)

pos = nx.spring_layout(G, seed=42)
edge_labels = nx.get_edge_attributes(G, 'weight')

plt.figure(figsize=(14, 6))

# Vẽ BFS
plt.subplot(1, 2, 1)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
if bfs_path:
    bfs_edges = list(zip(bfs_path, bfs_path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=bfs_edges, edge_color='red', width=3)
    nx.draw_networkx_nodes(G, pos, nodelist=bfs_path, node_color='orange', node_size=1200)
plt.title("BFS Path A → J")

# Vẽ DFS
plt.subplot(1, 2, 2)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
if dfs_path:
    dfs_edges = list(zip(dfs_path, dfs_path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=dfs_edges, edge_color='purple', width=3)
    nx.draw_networkx_nodes(G, pos, nodelist=dfs_path, node_color='orange', node_size=1200)
plt.title("DFS Path A → J")

plt.tight_layout()
plt.show()
````
